```python
from qiskit.quantum_info import *
from qiskit import *

import numpy as np

from HamOp import *
from Ham import *

from math import *
```

```python
# Parametri sistema

# Numero spin anello
n_spin = 2

# Periodicià anello
chiuso = True

# Costanti d'interazione [in unità di J]
J = [1,1,1]

# Parametro per la regolazione della velocità del processo di annealing
lam = 0.1

# Tempo in cui valutare la rappresentazione dell'crea_hamiltoniana
t_eval = 0.84275 / lam

# Funzione di drive per l'annealing
def drive(param, t):
#   Drive lineare
    ris = sin((param * t) * pi/2) ** (0.9)

    return ris

# Funzione di drive per l'annealing del termine a singolo corpo
def drive_sp(param, n, t):
    ris = tan(1) - tan((param * t) ** n)

    return ris
```

```python
# Creiamo la rappresentazione matriciale dell'Hamiltoniana tramite le interfacce standard

# Creazione Hamiltoniana
H_std = crea_hamiltoniana(n_spin, chiuso)

# Valutazione Hamiltoniana di Heisenberg del sistema
H_std_fin = H_std.valuta_tot(J, drive, lam, t_eval)
```

```python
# Creiamo la rappresentazione matriciale dell'Hamiltoniana tramite le interfacce operatoriali

# Creazione Hamiltoniana
H_Op = crea_hamiltonianaOp(n_spin, chiuso)

# Valutazione Hamiltoniana di Heisenberg del sistema
H_Op_fin = H_Op.valuta_tot(J, drive, lam, t_eval)
```

```python
# Calcolo norma differenza dei due operatori
np.linalg.norm(H_std_fin - H_Op_fin, 'fro')
```
